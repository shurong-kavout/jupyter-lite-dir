```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# 不发出警告
warnings.filterwarnings('ignore') 

# 导入 notebook 绘图模块
from bokeh.io import output_notebook
output_notebook()

# 导入图表绘制、图标展示模块
from bokeh.plotting import figure, show
```

```python
# demo1: 单折线图

from bokeh.models import ColumnDataSource
df = pd.DataFrame({
    'value': np.random.randn(100).cumsum()
})
dic = {
    'x': df.index.tolist(),
    'y': df['value'].tolist()
}
source = ColumnDataSource(data=dic)

p = figure(plot_width=600, plot_height=400)
p.line(
    x='x',
    y='y',
    source=source,
#     line_width=1,
#     line_alpha=0.8,
#     line_color='firebrick',
#     line_dash=[6, 3],
)

show(p)
```

```python
# demo2: 多折线图 - multi_line

df = pd.DataFrame({
    'value_1': np.random.randn(100).cumsum(),
    "value_2": np.random.randn(100).cumsum()
})
x1 = df['value_1'].index.tolist()
x2 = df['value_2'].index.tolist()
y1 = df['value_1'].tolist()
y2 = df['value_2'].tolist()

p = figure(plot_width=600, plot_height=400)
p.multi_line(
    xs=[x1, x2],
    ys=[y1, y2],
    color=["firebrick", "navy"],
    alpha=[0.8, 0.6],
    line_width=[1.5, 1.5],
)

show(p)
```
